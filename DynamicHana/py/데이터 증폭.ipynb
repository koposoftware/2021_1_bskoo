```python
import pandas as pd
import numpy as np
import openpyxl
from faker import Faker ## 샘플데이터 생성 라이브러리
fake = Faker('ko-KR')

import collections
from collections import OrderedDict
```

### t_member

```python
t_member = pd.read_csv('t_member.csv', encoding='cp949')
```

```python
temp = t_member
```

```python
temp
```

```python
for i in range(0,40):
    t_member = pd.concat([t_member,temp])
```

```python
t_member['ID']=''
t_member['ID']=t_member['ID'].apply(lambda x:fake.email()+str(fake.pyint(min_value=0, max_value=99999)))
```

```python
t_member['RRN']=''
t_member['RRN']=t_member['RRN'].apply(lambda x:str(fake.pyint(min_value=600000, max_value=991231))+"-"+str(fake.pyint(min_value=1111111, max_value=2399999)))
```

```python
t_member.reset_index(inplace=True)
t_member.reset_index(inplace=True)
```

```python
del t_member['index']
del t_member['MEMBER_NO']
t_member['level_0'] = t_member['level_0']+1
t_member = t_member.rename(columns={'level_0':'MEMBER_NO'})
```

```python
t_member
```

```python
t_member.to_csv('bigdata/t_member.csv')
```
